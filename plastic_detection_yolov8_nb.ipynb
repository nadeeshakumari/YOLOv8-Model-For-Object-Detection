```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8s.pt")
```

```python
from ultralytics import YOLO
from IPython.display import display, Image
```

```python
model.train(data="/kaggle/input/underwater-plastic-pollution-detection/underwater_plastics/data.yaml", epochs=50)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load and save with 300 DPI
img = Image.open("/kaggle/working/runs/detect/train/results.png")
img.save("/kaggle/working/high_res_results.png", dpi=(600, 600))  # Sets metadata

# Display in notebook with enhanced quality
plt.figure(dpi=300)
plt.imshow(img)
plt.axis('off')
plt.show()
```
